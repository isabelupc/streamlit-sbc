```python
titols = cases.llibre_recomanat.unique().tolist()
len(titols)
```

```python
cases[cases['llibre_recomanat'].isin([titols[0]])]['score'].mean()
```

```python
lectures = [len(cases[cases['llibre_recomanat'].isin([titols[i]])]) for i in range(90)]
```

```python
valoracions = [cases[cases['llibre_recomanat'].isin([titols[i]])]['score'].mean() for i in range(90)]
```

```python
ids = [i for i in range(1,91)]


# Create a DataFrame from the generated data
books = pd.DataFrame()
books['id_llibre'] = ids
books['titol'] = titols
books['autor'] = 'Autor'
books['num_lectors'] = lectures
books['valoracions'] = valoracions

books.to_csv('Books.csv', index=False, encoding='utf-8')

books
```
